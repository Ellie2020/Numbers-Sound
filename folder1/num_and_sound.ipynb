```python
#plotting the annoying continuos sound..called data :-)

import numpy
import matplotlib.pyplot as plt
data = numpy.random.uniform(low=-1,high=1,size=44100)
plt.plot(data)
plt.show()
```

```python
#Apply Fast Fourier Transform to the continuos signal..Here we get both real and imaginary part
from scipy.fftpack import fft, rfft
data = numpy.random.uniform(low=-1,high=1,size=44100)
data_frequency = np.fft.fft(data)
plt.plot(data_frequency)
```

```python
#Select the real part of the Fourier transform as we want the amplitude of 
#the sound in a specific frequency's.
#Signals phase (imagenery part of the FT) is used in image processing and audio digital filters 
#(e.g. sound distortion: different frequencies played with no delay)
#Apply Fast Fourier Transform to the continuos signal..Here we get just the real part that is what 
#really count for sound signals
from scipy.fftpack import fft, rfft
data_frequency_real = np.fft.rfft(data)
plt.plot(data_frequency_real)
```

```python
#generating a middle C (e.g. C4 in scientific pitch notation) with a frequency of 261.6 Hz,
#because of the note's position as the fourth C key on a standard 88 key piano keyboard.

import sounddevice as sd
sd.default.samplerate = 44100

time = 1

# Generate time of samples between 0 and time (= 1 sec)
samples = np.arange(44100 * time) / 44100

#Sinusoidal wave of frequency f has formula w(t) = A*sin(2*pi*f*t)
#Turning the bass on
frequency = 261.6
wave = 10000 * np.sin(2 * np.pi * frequency * samples)
sd.play(wave, blocking=True)

# Convert it to wav format (16 bits)
#wav_wave = np.array(wave, dtype=np.int16)

#sd.play(wav_wave, blocking=True)
```
